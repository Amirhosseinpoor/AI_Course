```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import Callback
import matplotlib.pyplot as plt
from tqdm.keras import TqdmCallback

# --- بازسازی داده به 2D ---
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# --- تعریف callback برای ذخیره خطا ---
class LossHistory(Callback):
    def __init__(self):
        super().__init__()
        self.train_loss = []
        self.val_loss = []

    def on_epoch_end(self, epoch, logs=None):
        self.train_loss.append(logs['loss'])
        self.val_loss.append(logs['val_loss'])

# --- تعریف تابع ساخت و آموزش مدل ---
def train_model_with_lr(lr_value):
    model = Sequential([
        Dense(16, input_shape=(X_train_flat.shape[1],), activation='relu'),
        Dense(1, activation='linear')
    ])

    optimizer = SGD(learning_rate=lr_value)
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

    history = LossHistory()

    model.fit(
        X_train_flat, y_train,
        validation_data=(X_test_flat, y_test),
        epochs=200,
        batch_size=32,
        callbacks=[history, TqdmCallback(verbose=0)],
        verbose=0
    )

    return history

# --- آموزش مدل‌ها برای سه نرخ یادگیری مختلف ---
learning_rates = [1e-3, 1e-8, 1e+0]
histories = {}
for lr in learning_rates:
    print(f"\nTraining model with learning rate = {lr}")
    histories[lr] = train_model_with_lr(lr)
```

```python
plt.figure(figsize=(14, 6))

for lr in learning_rates:
    plt.plot(histories[lr].val_loss, label=f'LR={lr} (Validation)')

plt.title('Validation Loss for Different Learning Rates')
plt.xlabel('Epoch')
plt.ylabel('MSE Loss')
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
import pandas as pd

# لیست نرخ‌های یادگیری
learning_rates = [1e-3, 1e-8, 1.0]
results = []

# تغییر epoch برای سریع‌تر شدن
EPOCHS = 50  # موقتاً کمتر برای تست سریع

for lr in learning_rates:
    print(f"\nTraining and evaluating model with learning rate = {lr}")

    # ساخت و آموزش مدل
    model = Sequential([
        Dense(16, input_shape=(X_train_flat.shape[1],), activation='relu'),
        Dense(1, activation='linear')
    ])

    optimizer = SGD(learning_rate=lr)
    model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

    model.fit(
        X_train_flat, y_train,
        epochs=EPOCHS,
        batch_size=32,
        verbose=0  # بی‌صدا برای سرعت بیشتر
    )

    # پیش‌بینی و ارزیابی
    y_pred = model.predict(X_test_flat, verbose=0).flatten()
    mse = mean_squared_error(y_test, y_pred)
    mae = mean_absolute_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    results.append((lr, mse, mae, r2))

# نمایش جدول نتایج
df_results = pd.DataFrame(results, columns=['Learning Rate', 'MSE', 'MAE', 'R2 Score'])
print(df_results)
```

```python
# ==== Import Libraries ====
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

# ==== Flatten Inputs if Needed ====
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# ==== Define Deeper Neural Network ====
deep_model = Sequential([
    Dense(64, input_shape=(X_train_flat.shape[1],), activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='linear')
])

# ==== Compile Model ====
optimizer = SGD(learning_rate=1e-3)
deep_model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# ==== Train Model ====
history = deep_model.fit(
    X_train_flat, y_train,
    validation_data=(X_test_flat, y_test),
    epochs=200,
    batch_size=32,
    verbose=1
)

# ==== Predict ====
y_pred = deep_model.predict(X_test_flat).flatten()

# ==== Evaluate Model ====
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Deep Neural Network Evaluation:")
print("MSE:", round(mse, 5))
print("MAE:", round(mae, 5))
print("R² Score:", round(r2, 5))

# ==== Plot Training and Validation Loss ====
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss (MSE)')
plt.plot(history.history['val_loss'], label='Validation Loss (MSE)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss - Deep Neural Network')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# ==== Import Libraries ====
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import mean_squared_error, mean_absolute_error, r2_score
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.callbacks import Callback

# ==== Flatten Inputs if Needed ====
X_train_flat = X_train.reshape(X_train.shape[0], -1)
X_test_flat = X_test.reshape(X_test.shape[0], -1)

# ==== Define Callback to Log Weights Properly ====
class WeightLogger(Callback):
    def __init__(self, layer_index=0, neuron_index=0):
        super().__init__()
        self.layer_index = layer_index
        self.neuron_index = neuron_index
        self.history = []

    def on_epoch_end(self, epoch, logs=None):
        # دقیقاً به سراغ وزن‌های لایه و نورون موردنظر می‌ریم
        weights = self.model.layers[self.layer_index].get_weights()[0]  # shape: (n_inputs, n_neurons)
        self.history.append(weights[:, self.neuron_index].copy())       # ذخیره وزن‌های ورودی به نورون iام

# ==== Define Model ====
deep_model = Sequential([
    Dense(64, input_shape=(X_train_flat.shape[1],), activation='relu'),
    Dense(32, activation='relu'),
    Dense(16, activation='relu'),
    Dense(1, activation='linear')
])

# ==== Compile Model ====
optimizer = SGD(learning_rate=0.01)  # نرخ یادگیری را کمی بیشتر گذاشتیم
deep_model.compile(optimizer=optimizer, loss='mse', metrics=['mae'])

# ==== Setup and Attach Callback ====
weight_logger = WeightLogger(layer_index=0, neuron_index=0)

# ==== Train Model ====
history = deep_model.fit(
    X_train_flat, y_train,
    validation_data=(X_test_flat, y_test),
    epochs=200,
    batch_size=32,
    verbose=1,
    callbacks=[weight_logger]
)

# ==== Predict & Evaluate ====
y_pred = deep_model.predict(X_test_flat).flatten()

mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("📊 Deep Neural Network Evaluation:")
print("MSE:", round(mse, 5))
print("MAE:", round(mae, 5))
print("R² Score:", round(r2, 5))

# ==== Plot Loss ====
plt.figure(figsize=(10, 5))
plt.plot(history.history['loss'], label='Train Loss (MSE)')
plt.plot(history.history['val_loss'], label='Validation Loss (MSE)')
plt.xlabel('Epoch')
plt.ylabel('Loss')
plt.title('Training and Validation Loss - Deep Neural Network')
plt.legend()
plt.grid(True)
plt.show()

# ==== Plot Weight Changes ====
weights_array = np.array(weight_logger.history)  # shape: (epochs, n_features)

plt.figure(figsize=(12, 6))
for i in range(min(5, weights_array.shape[1])):  # نمایش ۵ ویژگی اول
    plt.plot(weights_array[:, i], label=f'Feature {i} → Neuron 0')

plt.xlabel('Epoch')
plt.ylabel('Weight Value')
plt.title('Weight Change of Neuron 0 in First Hidden Layer')
plt.legend()
plt.grid(True)
plt.show()
```
